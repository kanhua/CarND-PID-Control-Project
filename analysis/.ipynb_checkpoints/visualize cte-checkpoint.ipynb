```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
data_dir="../cmake-build-debug/"
```

cte1: 0.2,0.0001,2

cte2: 0.1,0.0001,1

cte3: 0.3,0.0001,7 no twiddle

cte4: 0.2,0.0001,2 with twiddle


```python
fileset=[('testout.csv','P=0.2,I=0.0001,D=2'),
        ('testout2.csv','P=0.1,I=0.0001,D=1'),
        #('testout3.csv','P=0.3,I=0.0001,D=7'),
        ('testout5.csv','P=0.172965,I=0.00109726,D=1.75581'),
        ('testout6.csv','P=0.172965,I=0.00509726,D=1.75581')
        ]
```

```python
for file,label in fileset:
    infile=os.path.join(data_dir,file)
    data=np.loadtxt(infile)
    plt.plot(data[0:2000],label=label,alpha=0.7)
plt.legend()
plt.ylabel("CTE")
plt.xlabel("time")
plt.tight_layout()
plt.savefig("cte_result.png")
```
